### Load data

```python
import pandas as pd
import numpy as np
import re
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer,word_tokenize

df=pd.read_csv('./tweets_selfdrivingcars.csv')
df_text=df[['Tweet']]
df_text.head()
```

### Remove URL


```python
def remove_urls(text):
    #url_pattern = re.compile(r'https?://\S+|www\.\S+')
    url_pattern = re.compile(r'https?://[A-Za-z0-9./]+')
    return url_pattern.sub(r'', text)

for i in range(len(df_text['Tweet'])):
  df_text.loc[i, 'nourl_message'] = remove_urls(df_text.loc[i, 'Tweet'])
df_text = df_text.drop(columns='Tweet')

df_text.head()
```

### Remove @username

```python
def remove_username(text):
    return re.sub(r'RT @\w+: |@\w+', '', text)

for i in range(len(df_text['nourl_message'])):
  df_text.loc[i, 'nousername_message'] = remove_username(df_text.loc[i, 'nourl_message'])
df_text = df_text.drop(columns='nourl_message')

df_text.head()
```

### Replace emoji

```python
import demoji
for i in range(len(df_text['nousername_message'])):
  df_text.loc[i, 'emoji_message'] = demoji.replace_with_desc(df_text['nousername_message'][i]).replace(":"," ")
df_text = df_text.drop(columns='nousername_message')
df_text.head()
```

### Lowercasing

```python
for i in range(len(df_text['emoji_message'])):
  df_text.loc[i, 'low_message'] = df_text['emoji_message'][i].lower()
  #df_text['message']=df_text['message'].str.lower()
df_text = df_text.drop(columns='emoji_message')
df_text.head()
```

### Removing Extra Whitespaces

```python
def remove_whitespace(text):
    return  " ".join(text.split())

for i in range(len(df_text['low_message'])):
  df_text.loc[i, 'punc_message'] = remove_whitespace(df_text.loc[i, 'low_message'])
df_text = df_text.drop(columns='low_message')

#df_text['message']=df_text['message'].apply(remove_whitespace)
df_text.head()
```

### Removal of Tags


```python
def remove_tag(text):
    return re.sub(r'#\w*', '', text)
  
df_text['punc_message']=df_text['punc_message'].apply(remove_tag)
df_text.head()
```

Expand contractions in Text Processing

```python
import contractions
#df_text['message']=df_text['message'].str.lower()
#df_text['message']=contractions.fix(df_text['message'])
df_text['punc_message']=df_text['punc_message'].apply(contractions.fix)
df_text.head()
```

### Removing Punctuations

```python
def remove_punct(text):
    return re.sub(r'[^\w\s]','',text)

df_text['punc_message'] = df_text['punc_message'].apply(remove_punct)
df_text.head()
```

### Tokenization

```python
df_text['message']=df_text['punc_message'].apply(lambda X: word_tokenize(X))
df_text = df_text.drop(columns='punc_message')
df_text.head()
```

### Removing Frequent Words
##### Even after removing stopwords, some words occur in our text with very high frequency and yet may not add any value. 

```python
from nltk import FreqDist

def frequent_words(df):
    
    lst=[]
    for text in df.values:
        lst+=text[0]
    fdist=FreqDist(lst)
    return fdist.most_common(10)
frequent_words(df_text)
```

```python
freq_words = frequent_words(df_text)

lst = []
for a,b in freq_words:
    lst.append(b)

def remove_freq_words(text):
    
    result=[]
    for item in text:
        if item not in lst:
            result.append(item)
    
    return result
    
df_text['message']=df_text['message'].apply(remove_freq_words)
```

### Lemmatization
##### Lemmatization is converting the word to its base form or lemma by removing affixes from the inflected words. It helps to create better features for machine learning and NLP models hence it is an important preprocessing step.
